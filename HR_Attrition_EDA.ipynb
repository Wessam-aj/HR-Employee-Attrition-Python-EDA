```python
import pandas as pd
pd.set_option('display.max_columns',None)
df = pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head(6)
```

```python
df.groupby('Attrition')[['JobSatisfaction','RelationshipSatisfaction','WorkLifeBalance','DistanceFromHome','MonthlyIncome']].mean()
```

```python
pd.crosstab(df['OverTime'], df['Attrition'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# السطر الثاني: رسم عمود يمثل العمل الإضافي، وتلوينه بناءً على الاستقالة
sns.countplot(x='OverTime', hue='Attrition', data=df)

# السطر الثالث: وضع عنوان للرسمة
plt.title("Impact of OverTime on Attrition")

# السطر الرابع: إظهار الرسمة
plt.show()
```

```python
# رسم مخطط الصندوق للرواتب بناءً على الاستقالة
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set1')

# وضع عنوان احترافي
plt.title("Salary Distribution by Attrition Status")

# عرض الرسمة
plt.show()
```
